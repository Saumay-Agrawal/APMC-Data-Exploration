# Cleaning Dataset 1

### Loading the CMO MSP dataset into a dataframe

```python
df1 = pd.read_csv('./data/CMO_MSP_Mandi.csv')
df1.head()
```

### Basic exploration for finding out the column datatypes, occurence of null values, and statistical distribution of numerical values

```python
df1.info()
```

```python
df1.describe()
```

```python
for col in df1.columns:
    print('{}: {}'.format(col, df1[col].unique().size))
```

##### Upon checking out the number of unique values in each column, I could find out that for each commodity there could be the data for five years. However, since the total number of rows in dataset is 155 (which is less than 32*5), there could be some error in the commodity names. The fields containing text values mostly exhibit this behaviour where a small character difference creates two different entities.

```python
commodity_names = df1.commodity.unique()
commodity_names.sort()
print(commodity_names)
```

#### So from the above sorted list of unique commodity values present in the dataset, we can see that the commodity 'Sugarcane' is having two entries ie. "SUGARCANE" and "SUGAR-CANE". Also there is a lot of inconsistency in terms of upper and lower cases, which slows down the readibility of textual data. In order to rectify this, I have converted all the fields with string values to upper case. I have also stripped the values of leading and trailing whitespaces, though this dataset didn't require it. Following this, I have substituted the entries of "SUGAR-CANE" with "SUGARCANE" and corrected one addition error in commodity spelling.

```python
df1 = df1.applymap(lambda x: x.strip().upper() if (type(x)==str) else x)
df1.head()
```

```python
df1.commodity.replace(to_replace='SUGAR-CANE', value='SUGARCANE', inplace=True)
df1.commodity.replace(to_replace='SPILT GERRN GRAM', value='SPLIT GREEN GRAM', inplace=True)
commodity_names = df1.commodity.unique()
commodity_names.sort()
print(commodity_names)
```

```python
types = df1.Type.unique()
print(types)
```

#### Sorting the dataset for better readibility. This allowed me to see a trend of 5 year MSP data per commodity. This now alligns with the fact that there 155 rows, with 31 commodities, each having data for 5 years.

```python
df1.sort_values(by=['commodity', 'year'], inplace=True)
df1.head(155)
```

```python
df1[df1.isnull().any(axis=1)]
```

#### Though null values were visible in the previous result, I needed to get an exact count of null values. Upon viewing all the rows containing null values, I could see that most of commodities had missing MSP values for the years 2015 and 2016 only. This meant that MSP values for the same commodity were available for the previous years and forward filling of null values will suffice.

```python
df1.ffill()
```

#### Below I have plotted a simple line plot for MSPs of each commodity. This plot helped me in understanding that the price distribution of the commodities had a very wide range, which I should take care of while making further analysis. For eg, I couldn't mix the price values of potato and cashew as one is way more expensive than the other, and thus the price ranges would also vary vastly.

```python
sb.lineplot(data=df1, x='year', y='msprice', hue='commodity')
```

```python
df1.to_csv('./data/CMO_MSP_Mandi_cleaned.csv')
```

# Cleaning Dataset 2

```python
## load dataset
```

```python
df2 = pd.read_csv('./data/Monthly_data_cmo.csv', parse_dates=['date'])
df2.head()
```

```python
## basic exploration
```

```python
df2.info()
```

```python
df2.describe()
```

```python
for col in df2.columns:
    print('{}: {}'.format(col, df2[col].unique().size))
```

```python
df2.drop(['Year', 'Month', 'state_name'], axis=1, inplace=True)
df2.head()
```

#### I dropped the columns 'Year' and 'Month' in favour of having just one column 'date' to represent timestamps. If we need any additional information (like year and month) we can get it through the timestamp values. I have also dropped the 'state' name column as it was irrelevant for given dataset as no data is available for other states.

```python
commodities2 = df2.Commodity.unique()
commodities2.sort()
print(commodities2)
```

```python
df2 = df2.applymap(lambda x: x.strip().upper() if (type(x)==str) else x)
df2.head()
```

```python
df2.Commodity.replace(to_replace='SPILT GERRN GRAM', value='SPLIT GREEN GRAM', inplace=True)
df2.Commodity.replace(to_replace='BHAGAR/VARI', value='BHAGAR-VARI', inplace=True)
df2.Commodity.replace(to_replace='THYMOL/LOVAGE', value='THYMOL-LOVAGE', inplace=True)

commodities2 = df2.Commodity.unique()
commodities2.sort()
print(commodities2)
```

#### I have done the same procedure for cleaning the text values. Here many commodities were repeated due to case differences like 'CORIANDER ' and 'Coriander ', though leading space was the same. Converting to upper case solved this problem.

#### I have also replaced some commodity names as they created errors while saving the plots with commodity names in later analysis.

```python
sb.countplot(data=df2, x='Commodity')
```

```python
sb.countplot(data=df2, x='APMC')
```

#### I have used count plots to roughly estimate the number of entries for various APMC and Commodity clusters. Each APMC cluster will have further entries for each Commodities, and similarly each Commodity cluster will have further entries for each APMC.

```python
## checking for null values
```

```python
df2[df2.isnull().any(axis=1)]
```

#### There were no null values found in the dataset

#### For outlier detection, I have considered each Commodity individually. This is mainly because of the difference in the price ranges for different commodities. So the values for one commodity may become outliers for the other commodities.  I have plotted the box plots for the modal prices of a single commodity to show the presence of outliers and their removal. 

#### For removal of outliers, I have used the IQR (Inter Quartile Range) method. This removes the values before Q1 and after Q3 for a series. I have also plotted scatter plots for better visualisation of modal price values against the timestamps (with and without outliers). These scatter plots depict the price distribution much better than the boxplots. The scatter plots (with and without ouliers) were saved in a separate folder for all the commodities.

```python
df2.sort_values(by=['Commodity', 'date', 'APMC'], inplace=True)
```

```python
bajri = df2.loc[df2.Commodity=='BAJRI']
bajri.head()
```

```python
sb.boxplot(data=bajri, x='Commodity', y='modal_price')
```

```python
# Removing the outliers using IQR
q1 = bajri['modal_price'].quantile(.25)
q3 = bajri['modal_price'].quantile(.75)
mask = bajri['modal_price'].between(q1, q3, inclusive=True)
bajri_clean = bajri.loc[mask]
sb.boxplot(data=bajri_clean, x='Commodity', y='modal_price')
```

```python
fig = sb.scatterplot(data=bajri, x='date', y='modal_price').set(xlim=('2014-01-01', None))
```

```python
fig = sb.scatterplot(data=bajri_clean, x='date', y='modal_price').set(xlim=('2014-01-01', None))
```

```python
for crop in df2.Commodity.unique():
    cropdf = df2.loc[df2.Commodity==crop]
    plot = sb.scatterplot(data=cropdf, x='date', y='modal_price', hue='Commodity').set(xlim=('2014-01-01', None))
    plt.savefig('./visuals/outlier-handling/{}_outliers.png'.format(crop))
    plt.clf()
```

#### While performing seasonality detection, I found out that even though the null values were absent, many entries for minimum, maximum and modal prices contained zero. I tried to consider the modal price as the average of the maximum and minimum price values for substituting zero values. However, on a closer inspection, I found out that the modal values were skewed towards the minimum price a lot for each entry. Thus I substituted the zeros by NANs and then used forward filling for filling of missing values before performing the outlier removal to maintain the consistency of the dataset.

```python
# Printing crops having modal price 0
for crop in df2.Commodity.unique():
#     print('\n  For crop:', crop)
    dfcrop = df2.loc[df2.Commodity==crop]
    dfcrop = dfcrop.groupby('date').mean()
    if (dfcrop.loc[dfcrop.modal_price==0].shape[0]>0):
        print(crop)
```

```python
# Substituting 0s with NaNs
df2 = df2.replace(to_replace=0, value=np.nan)
df2.info()
```

```python
# Printing columns with null values
df2[df2.isnull().any(axis=1)]
```

```python
# Forward filling of NANs
df2 = df2.ffill()
df2[df2.isnull().any(axis=1)]
```

```python
# Performing IQR for each set of commodity.
dfclean = []
for commodity in df2.Commodity.unique():
    comdf = df2.loc[df2.Commodity==commodity]
    q1 = comdf['modal_price'].quantile(.25)
    q3 = comdf['modal_price'].quantile(.75)
    mask = comdf['modal_price'].between(q1, q3, inclusive=True)
    comdf_clean = comdf.loc[mask]
    dfclean.append(comdf_clean)

dfclean = pd.concat(dfclean)
```

```python
# Saving commodity wise scatter plots after outlier removal
for crop in dfclean.Commodity.unique():
    plot = sb.scatterplot(data=dfclean.loc[dfclean.Commodity==crop], x='date', y='modal_price', hue='Commodity').set(xlim=('2014-01-01', None))
    plt.savefig('./visuals/outlier-handling/{}_cleaned.png'.format(crop))
    plt.clf()
```

```python
# Saving cleaned file in csv format.
dfclean.to_csv('./data/Monthly_data_cmo_cleaned.csv')
```
