```python
q1 = df['modal_price'].quantile(.25)
q3 = df['modal_price'].quantile(.75)
mask = df['modal_price'].between(q1, q3, inclusive=True)
iqr = df.loc[mask]
iqr.shape
```

```python
iqr.shape
```

```python
dfclean = []
for commodity in df.Commodity.unique():
    comdf = df.loc[df.Commodity==commodity]
    q1 = comdf['modal_price'].quantile(.25)
    q3 = comdf['modal_price'].quantile(.75)
    mask = comdf['modal_price'].between(q1, q3, inclusive=True)
    iqr = comdf.loc[mask]
    print(commodity, comdf.shape, iqr.shape)
    dfclean.append(iqr)
```

```python
dfclean = pd.concat(dfclean)
dfclean.shape
```

```python
for crop in dfclean.Commodity.unique():
#     plot = sb.boxplot(data=dfclean.loc[dfclean.Commodity==crop], x='Commodity', y='min_price', hue='Commodity')
#     plt.savefig('./outlier-detection/outliers-removed/{}-min.png'.format(crop))
#     plt.clf()
    plot = sb.scatterplot(data=dfclean.loc[dfclean.Commodity==crop], x='date', y='modal_price', hue='Commodity')
    plot.set(xlim=('2014-01-01', None))
    plt.savefig('./outlier-detection/outliers-scatter/{}_cleaned.png'.format(crop))
    plt.clf()
#     plot = sb.boxplot(data=dfclean.loc[dfclean.Commodity==crop], x='Commodity', y='max_price', hue='Commodity')
#     plt.savefig('./outlier-detection/outliers-removed/{}-max.png'.format(crop))
#     plt.clf()
```

```python
bajri = df.loc[df.Commodity=='BAJRI']
bajri.sort_values(by=['date', 'APMC'], inplace=True)
bajri.head(50)
```

```python
bajri_mean = bajri.groupby(['date']).mean()
bajri_mean
```

```python
bajri_mean.plot(style='.-')
```

```python
bajri_resample = bajri_mean.resample('D').mean().interpolate(method='spline', order=2)
bajri_resample.plot(style='.-')
```

```python
bajri_mean.modal_price.plot()
```

```python
from statsmodels.tsa.stattools import adfuller
adfresult = adfuller(bajri_mean.modal_price)
adfresult
```

```python
ts = bajri_mean.modal_price.resample('D').mean().interpolate(method='spline', order=2)
ts.plot()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
mulresult = seasonal_decompose(bajri_mean.modal_price, model='multiplicative', freq=12)
fig = mulresult.plot()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
addresult = seasonal_decompose(bajri_mean.modal_price, model='additive', freq=12)
fig = addresult.plot()
```

```python
from statsmodels.tsa.stattools import acf

acfmul = acf(mulresult.resid)
acfadd = acf(addresult.resid)
if(sum(acfmul)<=sum(acfadd)):
    print('multiplicative')
else:
    print('additive')
```

```python
## deseasonalised data
desdata = addresult.observed - addresult.seasonal
desdata.plot()
```

```python
bajri_mean['deseasonalised'] = desdata
bajri_mean[['modal_price', 'deseasonalised']].plot()
```

```python
def decide(series, freq):
    mulresult = seasonal_decompose(series, model='multiplicative', freq=12)
    addresult = seasonal_decompose(series, model='additive', freq=12)
    acfmul = acf(mulresult.resid)
    acfadd = acf(addresult.resid)
    if(sum(acfmul)<=sum(acfadd)):
        return 'multiplicative'
    else:
        return 'additive'
```

```python
for crop in df.Commodity.unique():
    print('/n/n For crop:', crop)
    dfcrop = df.loc[df.Commodity==crop]
    dfcrop = dfcrop.groupby('date').mean()
    if (dfcrop.shape[0]<24):
        print(dfcrop.shape[0] , 'rejected')
    else:
        print(decide(dfcrop.modal_price, 12))
```
